# Using the model

```python
from pathlib import Path
import cv2
from ultralytics import YOLO

model = YOLO('runs/detect/training on 400 photos/weights/best.pt')

image_path = str(Path.cwd() / 'images' / 'test1.png')
img = cv2.imread(image_path)


results = model(img, verbose=False, conf=0.4)

annotated_image = results[0].plot(
    masks=False,    
    line_width=1,   
    labels=True,     
    boxes=True
)

output_path = '1.jpg'
cv2.imwrite(output_path, annotated_image)
```

# TRAIN

```python
import torch
from ultralytics import YOLO

model = YOLO("yolo11m.pt")

results = model.train(
    data="dataset/data.yaml",
    amp=True,
    epochs=200, 
    imgsz=736,
    device=0,
    batch=12, 
    model="yolo11m.pt",
    augment=True,
    mosaic=1.0,
    mixup=0.3,
    copy_paste=0.2,
    degrees=15,
    scale=0.7,
    fliplr=0.5,
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4
)
```
